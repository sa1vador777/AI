Rice (Cammeo and Osmancik)

Всего было сделано 3810 изображений рисовых зерен для двух видов, обработано и сделаны выводы о характеристиках. Для каждого зерна риса получено 7 морфологических признаков.

Признаки:

|Имя переменной|Роль|Тип|Описание|Единицы|
|--------------|----|---|--------|-------|
|Area|Особенность|Целое число|Возвращает количество пикселей в границах рисового зерна.|пикселей|
|Perimeter|Особенность|Непрерывный|Вычисляет длину окружности, вычисляя расстояние между пикселями вокруг границ рисового зерна.|пикселей|
|Major_Axis_Length|Особенность|Непрерывный|Самая длинная линия, которую можно провести на рисовом зерне, т. е. расстояние по главной оси||
|Minor_Axis_Length|Особенность|Непрерывный|Самая короткая линия, которую можно провести на рисовом зерне, т. е. малое расстояние по оси||
|Eccentricity|Особенность|Непрерывный|Он измеряет, насколько округлен эллипс, имеющий те же моменты, что и рисовое зерно.||
|Convex_Area|Особенность|Целое число|Возвращает количество пикселей наименьшей выпуклой оболочки области, образованной рисовым зерном.||
|Extent|Особенность|Непрерывный|Возвращает соотношение области, образованной рисовым зерном, к ограничивающей рамке.||
|Class|Цель|Двоичный|Каммео и Османчик||

- Форма набора данных: количество элементов набора, количество признаков, количество пропущенных значений, среднее значениеотдельных признаков, максимальные и минимальные значения отдельных признаков и прочие  показатели.Предположения,  которые  можно  сделать,  проведя первичный анализ.

```python
import numpy as np
import pandas as pd

data_path = 'rice.data'

dt = np.dtype("f8, f8, f8, f8, f8, f8, f8, U30")
data = pd.read_table(data_path, delimiter=',', header=None, names=['Area', 'Perimeter', 'Major_Axis_Length',
                                                                     'Minor_Axis_Length', 'Eccentricity',
                                                                     'Convex_Area', 'Extent', 'Class'])
print(f"Форма набора данных: {data.shape}\n")
print(f"Количество пропущенных значений:\n{data.isnull().sum()}\n")
print(f"\nОписательная статистика:\n{data.describe()}")
```

- Графические   представления,   позволяющие   судить   о неоднородности  исследуемого  набора  данных.  Построение  графиков желательно произвести по нескольким проекциям.

```python
import matplotlib.pyplot as plt


# Загрузка данных из файла "rice.data"
data = np.genfromtxt(data_path, delimiter=",", dtype=dt)

set_cammeo = [i for i in data if i[7] == 'Cammeo']
set_osmancik = [i for i in data if i[7] == 'Osmancik']

# Создаем списки для каждого из признаков
area_cammeo = [i[0] for i in set_cammeo]  # Область
perimeter_cammeo = [i[1] for i in set_cammeo]  # Периметр
major_axis_length_cammeo = [i[2] for i in set_cammeo]  #
minor_axis_length_cammeo = [i[3] for i in set_cammeo]  # Minor_Axis_Length
eccentricity_cammeo = [i[4] for i in set_cammeo]  # Эксцентриситет
convex_area_cammeo = [i[5] for i in set_cammeo]  # Выпуклая_область
degree_cammeo = [i[6] for i in set_cammeo]  # Степень

area_osmancik = [i[0] for i in set_osmancik]  # Область
perimeter_osmancik = [i[1] for i in set_osmancik]  # Периметр
major_axis_length_osmancik = [i[2] for i in set_osmancik]  #
minor_axis_length_osmancik = [i[3] for i in set_osmancik]  # Minor_Axis_Length
eccentricity_osmancik = [i[4] for i in set_osmancik]  # Эксцентриситет
convex_area_osmancik = [i[5] for i in set_osmancik]  # Выпуклая_область
degree_osmancik = [i[6] for i in set_osmancik]  # Степень

# Строим графики
plt.figure(figsize=(10, 8))

plt.subplot(3, 2, 1)
plt.scatter(area_cammeo, perimeter_cammeo, c='r', marker='o', label='Cammeo')
plt.scatter(area_osmancik, perimeter_osmancik, c='b', marker='^', label='Osmancik')
plt.xlabel('Область')
plt.ylabel('Периметр')
plt.legend()

plt.subplot(3, 2, 2)
plt.scatter(major_axis_length_cammeo, minor_axis_length_cammeo, c='r', marker='o', label='Cammeo')
plt.scatter(major_axis_length_osmancik, minor_axis_length_osmancik, c='b', marker='^', label='Osmancik')
plt.xlabel('Длинна главной оси')
plt.ylabel('Длинна малой оси')
plt.legend()

plt.subplot(3, 2, 3)
plt.scatter(eccentricity_cammeo, convex_area_cammeo, c='r', marker='o', label='Cammeo')
plt.scatter(eccentricity_osmancik, convex_area_osmancik, c='b', marker='^', label='Osmancik')
plt.xlabel('Эксцентриситет')
plt.ylabel('Выпуклая область')
plt.legend()

plt.subplot(3, 2, 4)
plt.scatter(degree_cammeo, area_cammeo, c='r', marker='o', label='Cammeo')
plt.scatter(degree_osmancik, area_osmancik, c='b', marker='^', label='Osmancik')
plt.xlabel('Степень')
plt.ylabel('Область')
plt.legend()

plt.tight_layout()
plt.show()
```
